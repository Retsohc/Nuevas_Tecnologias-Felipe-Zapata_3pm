```python
import pandas as pd
```

```python
datos_vivienda = pd.read_csv('alquileres_viviendas.csv', sep=';')
```

```python
datos_vivienda
```

```python
# Seleccione tipo 'Departamento'

sel_departamento = datos_vivienda['Tipo'] == 'Departamento'
n1 = datos_vivienda[sel_departamento].shape[0]
n1
```

```python
seleccion = (datos_vivienda['Tipo'] == 'Casa') | (datos_vivienda['Tipo'] == 'Habitación')
n2 = datos_vivienda[seleccion].shape[0]
n2
```

```python
sel_area = (datos_vivienda['Area'] >= 60) & (datos_vivienda['Area'] <= 100)
n3 = datos_vivienda[sel_area].shape[0]
n3
```

```python
# Cuartos >= 4 y alquileres <= 2000

selec_cuartos_aquiler = (datos_vivienda['Cuartos'] >= 4) & (datos_vivienda['Valor'] <= 2000)
n4 = datos_vivienda[selec_cuartos_aquiler].shape[0]
n4
```

```python
print(f"El número de departamentos es de {n1}")
print(f"El número de casas y habitaciónes encontradas es de {n2}")
print(f"El número de viviendas con un area entre 60 y 100 metros cuadrados es de {n3}")
print(f"El número de viviendas que tengan al menos 4 habitaciónes y cuesten menos de 2000 son {n4}")
```

```python
datos_vivienda.to_csv('datos_vivienda.csv', sep=';')
```
